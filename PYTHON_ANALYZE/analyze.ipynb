```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
```

# Wpływ szerokości bitowej na czas algorytmu

```python
df_8 = pd.read_csv('bit_size_10k_iter/8_bit', delimiter='\t')
df_16 = pd.read_csv('bit_size_10k_iter/16_bit', delimiter='\t')
df_32 = pd.read_csv('bit_size_10k_iter/32_bit', delimiter='\t')
df_64 = pd.read_csv('bit_size_10k_iter/64_bit', delimiter='\t')

df = pd.DataFrame(data={
    'Szerokość liczby': [
        8,
        16,
        32,
        64
    ],
    'Średni czas CPU [ms]': [
        df_8['CPU'].mean(),
        df_16['CPU'].mean(),
        df_32['CPU'].mean(),
        df_64['CPU'].mean()
    ],
    'Odchylenie standardowe CPU [ms]': [
        df_8['CPU'].std(),
        df_16['CPU'].std(),
        df_32['CPU'].std(),
        df_64['CPU'].std()
    ],
    'Średni czas GPU [ms]': [
        df_8['GPU'].mean(),
        df_16['GPU'].mean(),
        df_32['GPU'].mean(),
        df_64['GPU'].mean()
    ],
    'Odchylenie standardowe GPU [ms]': [
        df_8['GPU'].std(),
        df_16['GPU'].std(),
        df_32['GPU'].std(),
        df_64['GPU'].std()
    ]
})

df
```

```python
plt.plot(df['Szerokość liczby'], df['Średni czas CPU [ms]'], marker='o', label='Średni czas CPU')
plt.plot(df['Szerokość liczby'], df['Średni czas GPU [ms]'], marker='o', label='Średni czas GPU')

# Setting x-axis ticks to match "Szerokość liczby"
plt.xticks(df['Szerokość liczby'])

# Adding labels, title, and legend
plt.title('Średni czas CPU i GPU w zależności od szerokości liczby (10000 iteracji)')
plt.xlabel('Szerokość liczby w bitach')
plt.ylabel('Średni czas [ms]')
plt.legend()
plt.grid(True)
plt.show()
plt.close()
```

# Wydajność testowania wielu liczb jednocześnie


```python
df_100 = pd.read_csv('multi_number_file_1k_iter/100_num', delimiter='\t')
df_500 = pd.read_csv('multi_number_file_1k_iter/500_num', delimiter='\t')
df_1000 = pd.read_csv('multi_number_file_1k_iter/1000_num', delimiter='\t')
df_2500 = pd.read_csv('multi_number_file_1k_iter/2500_num', delimiter='\t')
df_5000 = pd.read_csv('multi_number_file_1k_iter/5000_num', delimiter='\t')
df_10000 = pd.read_csv('multi_number_file_1k_iter/10000_num', delimiter='\t')
df_100k = pd.read_csv('multi_number_file_1k_iter/100k_num_only_gpu', delimiter='\t')

df = pd.DataFrame(data={
    'Liczb w jednym pliku': [
        100,
        500,
        1000,
        2500,
        5000,
        10000,
        100000
    ],
    'Średni czas CPU [ms]': [
        df_100['CPU'].mean(),
        df_500['CPU'].mean(),
        df_1000['CPU'].mean(),
        df_2500['CPU'].mean(),
        df_5000['CPU'].mean(),
        df_10000['CPU'].mean(),
        numpy.nan,
    ],
    'Odchylenie standardowe CPU [ms]': [
        df_100['CPU'].std(),
        df_500['CPU'].std(),
        df_1000['CPU'].std(),
        df_2500['CPU'].std(),
        df_5000['CPU'].std(),
        df_10000['CPU'].std(),
        numpy.nan,
    ],
    'Średni czas GPU [ms]': [
        df_100['GPU'].mean(),
        df_500['GPU'].mean(),
        df_1000['GPU'].mean(),
        df_2500['GPU'].mean(),
        df_5000['GPU'].mean(),
        df_10000['GPU'].mean(),
        df_100k['GPU'].mean(),
    ],
    'Odchylenie standardowe GPU [ms]': [
        df_100['GPU'].std(),
        df_500['GPU'].std(),
        df_1000['GPU'].std(),
        df_2500['GPU'].std(),
        df_5000['GPU'].std(),
        df_10000['GPU'].std(),
        df_100k['GPU'].std(),
    ]
})

df
```

```python
plt.plot(df['Liczb w jednym pliku'], df['Średni czas CPU [ms]'], marker='o', label='Średni czas CPU')
plt.plot(df['Liczb w jednym pliku'], df['Średni czas GPU [ms]'], marker='o', label='Średni czas GPU')

plt.title('Średni czas CPU i GPU w zależności od liczb w jednym pliku (1000 iteracji)')
plt.xlabel('Liczb w jednym pliku')
plt.ylabel('Średni czas [ms]')
plt.legend()
plt.grid(True)
plt.show()
plt.close()
```
